```python
import pandas as pd
import sys
from geopy.distance import geodesic
```

```python
data = pd.read_csv('./data/merge_hw_RouteStation.csv')
data.head()
```

```python
data[data['bus_line_no']=='20-2(용포화석지)']
```

```python
# 모든 노선 중복도 확인

count = 0
for i in data['pr_station_id'].unique():
    count += 1
    
    temp = data[data['pr_station_id'] == i]
    name = temp['bus_line_no'].unique()
    
    temp.dropna(subset=['WGS84위도'], inplace = True)
    temp = temp.sort_values(by = 'bus_line_no_seq')
    temp = temp.iloc[0:round(len(temp)/2)][['WGS84위도','WGS84경도']]
    
    total = 0
    for j in range(len(temp)-1):
        O = temp.iloc[j, 0], temp.iloc[j, 1]
        D =  temp.iloc[j+1, 0], temp.iloc[j+1, 1]
        total += geodesic(O, D).km
    
    Origin = temp.iloc[0, 0], temp.iloc[0, 1]
    Destination = temp.iloc[len(temp)-1, 0], temp.iloc[len(temp)-1, 1]
    
    distance = geodesic(Origin, Destination).km
    
    sinuosity = total/distance
    
    print(i, name, sinuosity, end = ' ')
    
    if count % 2 == 0:
        print('')
```
